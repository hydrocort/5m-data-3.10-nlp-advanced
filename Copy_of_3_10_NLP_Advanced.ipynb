# Deep Learning for NLP

Deep learning for NLP involves using these neural networks to automatically discover the representations of words and the structure of language, which is necessary for tasks like translation, sentiment analysis, and question answering.

Key Advantages of Deep Learning in NLP:

- **Automatic Feature Extraction**: Unlike traditional models that require handcrafted features, deep learning models can learn to represent language data automatically.
- **Handling of Sequential Data**: Deep learning models, especially Recurrent Neural Networks (RNNs), are inherently suited for sequential data such as text.
- **Scalability**: Deep learning models can scale with data and compute power, often leading to improved performance with more data.
- **Contextual Understanding**: With architectures like the Transformer, deep learning models can understand the context and relationships within text.

## Common Tasks in NLP

Let's go through some of the common tasks in NLP.

### 1. Sequence Modeling
- **Language Modeling**: Predicting the probability distribution of the next word in a sequence given the previous words.
- **Text Generation**: Automatically generating text that is coherent and contextually relevant.

### 2. Text Classification
- **Sentiment Analysis**: Determining whether a piece of text expresses a positive, negative, or neutral sentiment.
- **Spam Detection**: Identifying and filtering out unsolicited and unwanted messages.
- **Topic Classification**: Assigning topics or categories to text based on its content.

### 3. Information Extraction
- **Named Entity Recognition (NER)**: Locating and classifying named entities in text into predefined categories such as names of persons, organizations, locations, etc.
- **Relation Extraction**: Identifying relationships between named entities.
- **Event Extraction**: Recognizing events and their corresponding attributes from text.

### 4. Sequence-to-Sequence Tasks
- **Machine Translation**: Translating text from one language to another.
- **Summarization**: Producing a concise and fluent summary while preserving key information and overall meaning.
- **Question Answering**: Generating a response to a query in natural language.

### 5. Parsing
- **Syntactic Parsing**: Analyzing the grammatical structure of a sentence to derive its parse tree.
- **Semantic Parsing**: Determining the meaning of a sentence by mapping it to a formal representation, such as a logical form or a query against a database.

### 6. Speech Processing
- **Speech Recognition**: Converting spoken language into text.
- **Text-to-Speech (TTS)**: Generating natural-sounding speech from text.

### 7. Natural Language Understanding (NLU)
- **Intent Classification**: Identifying the intention behind a given piece of text, commonly used in conversational AI.
- **Slot Filling**: Extracting semantic entities that correspond to specific slots from text, also used in conversational AI.

### 8. Natural Language Generation (NLG)
- **Dialogue Systems**: Creating systems that can converse with humans in natural language.
- **Automated Storytelling**: Generating narrative content for stories or reports.

### 9. Text Similarity and Clustering
- **Semantic Textual Similarity**: Measuring the degree of semantic equivalence between two pieces of text.
- **Document Clustering**: Grouping text documents into clusters based on their content similarity.

### 10. Coreference Resolution
- **Pronoun Resolution**: Determining the entities that pronouns refer to within a text.
- **Entity Linking**: Disambiguating entities to knowledge base identifiers.

## Transition from Traditional Machine Learning to Deep Learning Models

As we learnt in the previous lesson, traditional machine learning models for NLP, such as Naive Bayes, rely heavily on the process of feature engineering. This process involves transforming raw data into an appropriate set of features to help the models understand the language structure, e.g. TF-IDF.

### Limitations of Traditional Models:

- **Sparsity**: Traditional text representations like bag-of-words result in sparse vectors that are not efficient for capturing the semantic meaning of words.
- **Feature Engineering**: Crafting features requires domain expertise and is often labor-intensive and time-consuming.
- **Contextual Information**: Capturing context and word order is challenging with traditional models.

### Emergence of Deep Learning Models:

Deep learning models introduced end-to-end learning, where raw data can be fed into the model, and the model itself learns the features that are most predictive for the task.

- **Word Embeddings**: Deep learning introduced word embeddings (e.g., Word2Vec, GloVe), which are dense vector representations of words capturing semantic meaning.
- **Sequence Models**: RNNs and their variants like LSTMs and GRUs were developed to handle sequences and capture temporal dependencies in text.
- **Attention Mechanisms**: Attention allows models to focus on different parts of the input sequence when predicting an output, improving performance on tasks like machine translation.
- **Transformer Models**: The Transformer architecture, which relies solely on attention without recurrence, has achieved state-of-the-art results in many NLP tasks.

# Recurrent Neural Networks (RNNs)

## Theoretical Foundation

Recurrent Neural Networks (RNNs) are a powerful class of neural networks designed to handle sequential data. Unlike traditional neural networks, which assume that all inputs (and outputs) are independent of each other, RNNs are characterized by their "memory" as they process sequences of data. They can use their internal state (memory) to process variable-length sequences of inputs.

The defining feature of an RNN is the ability to form a directed cycle between nodes, which allows information to persist. In the context of sequences, this means that the network can retain information about previous elements in the sequence to influence the processing of the current element.

![rnn](https://www.researchgate.net/profile/Weijiang-Feng/publication/318332317/figure/fig1/AS:614309562437664@1523474221928/The-standard-RNN-and-unfolded-RNN.png)

In mathematical terms, at each time step $ t $, the hidden state $ h_t $ of the RNN is updated by:

$$ h_t = \sigma(W h_{t-1} + U x_t + b_h) $$
$$ o_t = V h_t + b_o $$

where:
- $ h_t $ is the hidden state at time $ t $,
- $ x_t $ is the input at time $ t $,
- $ o_t $ is the output at time $ t $,
- $ W $ is the weight matrix for the connections between the hidden units (recurrent weights),
- $ U $ is the weight matrix for the connections between the input and the hidden units,
- $ V $ is the weight matrix for the connections between the hidden units and the output,
- $ b_h $ and $ b_o $ are the bias terms for the hidden and output layers, respectively,
- $ \sigma $ is the activation function, typically tanh or ReLU.

## Limitations

Despite their power, RNNs suffer from a few limitations:

### Vanishing Gradient Problem

During backpropagation, RNNs suffer from the vanishing gradients problem, where gradients can become extremely small, effectively preventing the weights from changing their value. This makes it difficult for the RNN to learn long-range dependencies within the sequence.

The problem arises due to the nature of the backpropagation through time (BPTT) algorithm, where gradients are computed as products of derivatives. If these derivatives are small, the product can rapidly approach zero as the sequence gets longer, which leads to vanishing gradients.

### Exploding Gradient Problem

Conversely, RNNs can also suffer from the exploding gradients problem, where gradients can grow exponentially instead of diminishing, causing large updates to the weights and leading to an unstable network.

## Applications of RNNs

RNNs are well-suited for various NLP tasks, including but not limited to:

- **Language Modeling**: RNNs can predict the probability of the next word in a sentence given the previous words, which is fundamental in generating text.
- **Text Generation**: By sampling from the language model, RNNs can generate coherent sequences of text.
- **Machine Translation**: RNNs can be used to translate text from one language to another by modeling the conditional probability of the target sequence given the source sequence.
- **Speech Recognition**: RNNs are used to convert spoken language into text by modeling sequences of audio frames and predicting transcriptions.

```python
# 2. Tokenizer (basic English like torchtext's)
def tokenizer(text):
    return re.findall(r'\b\w+\b', text.lower())

tokenizer('HELLO!! this is a test text')
```

```python
# 3. Build vocabulary from training data
def build_vocab(data_iter, min_freq=1):
    counter = Counter()
    for item in data_iter:
        counter.update(tokenizer(item['sentence']))
    vocab = {'<unk>': 0}
    for word, freq in counter.items():
        if freq >= min_freq:
            vocab[word] = len(vocab)
    return vocab

vocab = build_vocab(dataset['train'])
vocab_size = len(vocab)
```

```python
def encode(text):
    return [vocab.get(token, vocab['<unk>']) for token in tokenizer(text)]

# 4. Process all splits
def data_process(split_data):
    encoded = [torch.tensor(encode(item['sentence']), dtype=torch.long) for item in split_data]
    return torch.cat([seq for seq in encoded if seq.numel() > 0])

train_data = data_process(dataset['train'])
val_data = data_process(dataset['validation'])
test_data = data_process(dataset['test'])
```

```python
train_data
```

```python
class TextDataset(Dataset):
    def __init__(self, text, seq_length):
        self.text = text
        self.seq_length = seq_length

    def __len__(self):
        return len(self.text) - self.seq_length

    def __getitem__(self, index):
        return (self.text[index:index+self.seq_length],
                self.text[index+1:index+self.seq_length+1])

# Define the sequence length and batch size
seq_length = 30 #20-50
batch_size = 64

# Create the dataset and data loader for training data
train_dataset = TextDataset(train_data, seq_length)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

```python
# Define the RNN model
class RNNModel(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size, num_layers=1):
        super(RNNModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_size)
        self.rnn = nn.RNN(embed_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)
        self.hidden_size = hidden_size
        self.num_layers = num_layers

    def forward(self, x, hidden):
        x = self.embedding(x)
        out, hidden = self.rnn(x, hidden)
        out = self.fc(out)
        return out, hidden

    def init_hidden(self, batch_size):
        return torch.zeros(self.num_layers, batch_size, self.hidden_size)

# Hyperparameters
vocab_size = len(vocab)  # size of the vocabulary
embed_size = 64          # embedding dimension
hidden_size = 128        # RNN hidden size

# Create the model, loss function, and optimizer
model = RNNModel(vocab_size, embed_size, hidden_size)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
# Switch to GPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
device
```

```python
# Train function with data loader
def train(model, data_loader, epochs):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for inputs, targets in data_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            # Initialize hidden state for each batch
            hidden = model.init_hidden(inputs.size(0)).to(device)
            optimizer.zero_grad()
            output, hidden = model(inputs, hidden)
            loss = criterion(output.view(-1, vocab_size), targets.view(-1))
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f'Epoch {epoch+1}/{epochs}, Loss: {total_loss / len(data_loader)}')
```

```python
train(model, train_loader, 10)
```

```python
itos = {idx: word for word, idx in vocab.items()}
def generate_text(model, vocab,itos, start_string, generation_length=100, temperature=1.0):
    model.eval()
    device = next(model.parameters()).device

    tokens = start_string.lower().split()
    input_ids = torch.tensor(
        [[vocab.get(token, vocab['<unk>']) for token in tokens]],
        dtype=torch.long
    ).to(device)

    hidden = model.init_hidden(1).to(device)
    words = tokens.copy()

    with torch.no_grad():
        for _ in range(generation_length):
            output, hidden = model(input_ids, hidden)
            last_logits = output[0, -1] / temperature
            probs = torch.nn.functional.softmax(last_logits, dim=0).cpu().numpy()
            next_id = np.random.choice(len(probs), p=probs)
            next_word = itos[next_id]
            words.append(next_word)

            input_ids = torch.tensor([[next_id]], dtype=torch.long).to(device)

    return ' '.join(words)
```

# Long Short-Term Memory (LSTM)

## Introduction to LSTMs and Their Architecture

Long Short-Term Memory networks (LSTMs) are a special kind of Recurrent Neural Network (RNN) capable of learning long-term dependencies. Introduced by Hochreiter & Schmidhuber in 1997, LSTMs have been developed to deal with the vanishing gradient problem that can occur in traditional RNNs.

An LSTM unit is composed of a cell, an input gate, an output gate, and a forget gate. The cell is responsible for keeping track of the dependencies between the elements in the input sequence. The gates regulate the flow of information into and out of the cell.

![lstm](https://databasecamp.de/wp-content/uploads/lstm-architecture-1024x709.png)

The operations within an LSTM unit can be summarized by the following equations:

- Forget Gate:
$$ f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f) $$

- Input Gate:
$$ i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i) $$
$$ \tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C) $$

- Cell State Update:
$$ C_t = f_t * C_{t-1} + i_t * \tilde{C}_t $$

- Output Gate:
$$ o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o) $$
$$ h_t = o_t * \tanh(C_t) $$

Here, $ \sigma $ represents the sigmoid function, and $ \tanh $ is the hyperbolic tangent function. The subscript $ t $ indexes the time step.

## How LSTMs Mitigate the Vanishing Gradient Problem

The vanishing gradient problem occurs during training of RNNs, where gradients can become very small exponentially quickly, as a function of the number of layers. Since the gradient is used to update the weights, if the gradient is too small, the weights cannot be updated effectively, and the RNN cannot learn long-range dependencies.

LSTMs mitigate the vanishing gradient problem through their gate structure and by maintaining a separate cell state over time. The cell state can carry relevant information throughout the processing of the sequence, so that even information from the early steps can have an impact on the later steps, without being multiplied repeatedly by potentially small weights.

# Gated Recurrent Units (GRU)

## Introduction to GRUs and Their Architecture

Gated Recurrent Units (GRUs) are a variant of the LSTM designed to adaptively capture dependencies of different time scales. Introduced by Cho et al. in 2014, GRUs are similar to LSTMs, but with a simplified structure. They combine the input and forget gates into a single "update gate" and also merge the cell state and hidden state.

![gru](https://pub.mdpi-res.com/ijgi/ijgi-09-00635/article_deploy/html/images/ijgi-09-00635-g001-550.jpg?1604379274)

The architecture of a GRU can be described by the following equations:

- Update Gate:
$$ z_t = \sigma(W_z \cdot [h_{t-1}, x_t]) $$

- Reset Gate:
$$ r_t = \sigma(W_r \cdot [h_{t-1}, x_t]) $$

- Candidate Hidden State:
$$ \tilde{h}_t = \tanh(W \cdot [r_t * h_{t-1}, x_t]) $$

- Final Hidden State:
$$ h_t = (1 - z_t) * h_{t-1} + z_t * \tilde{h}_t $$

The GRU has two gates: the update gate $ z_t $ decides how much of the past information to pass along to the future, and the reset gate $ r_t $ decides how much of the past information to forget.

## Comparison of GRUs with LSTMs

While both GRUs and LSTMs are designed to remember information for long periods, there are some differences in their architecture and performance:

- GRUs are simpler and have fewer parameters than LSTMs because they don't have an output gate and they combine the cell state and hidden state.
- LSTMs have an additional "cell state" which can maintain information in memory for long periods, independent of the hidden state. This may give LSTMs an advantage in problems where this capability is crucial.
- GRUs directly expose the entire hidden content without any control, whereas LSTMs control the exposure of the memory content through the output gate.
- In practice, GRUs tend to be faster to train than LSTMs due to their simpler structure, and they may perform better on smaller datasets. However, LSTMs might outperform GRUs on more complex tasks or datasets with longer sequences.

Choosing between GRUs and LSTMs often depends on the specific application and the dataset. In some cases, the performance difference may be minimal, and the choice may come down to computational efficiency.

# Attention Mechanisms

## Motivation for Attention in Sequence Modeling

Sequence modeling tasks, such as machine translation, often require the processing of input sequences (e.g., a sentence in a source language) to generate output sequences (e.g., a translated sentence in a target language). Traditional sequence-to-sequence models, like RNNs, encode the input sequence into a fixed-size context vector, which is then used to generate the output sequence. However, this approach has limitations, especially when dealing with long sequences. The fixed-size context vector becomes a bottleneck as it must capture all the information from the input sequence, regardless of its length.

Attention mechanisms address this limitation by allowing the model to automatically focus on different parts of the input sequence during the generation of each word in the output sequence. This selective focus enables the model to consider only the relevant information at each step of the output generation, alleviating the burden on the context vector to hold the entire sequence's information.

## How Attention Improves Model Performance

Attention mechanisms improve model performance by providing a dynamic context that adapts based on the output sequence. Instead of using a single context vector for the entire output sequence, attention uses a weighted combination of the input sequence's hidden states, where the weights reflect the relevance of each input element to the current output element.

The core idea of attention can be formulated as a weighted sum of values, where the weights are computed by a compatibility function of the query (from the decoder) and the keys (from the encoder):

$$ \text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V $$

Here, $ Q $, $ K $, and $ V $ stand for the query, key, and value, respectively, and $ d_k $ is the scaling factor (dimensionality of the keys).

![attention](https://upload.wikimedia.org/wikipedia/commons/thumb/8/81/Attention-qkv.png/700px-Attention-qkv.png)

This mechanism allows the model to learn what to attend to based on the data, leading to better handling of long-range dependencies and improved overall performance on tasks like machine translation, text summarization, and question answering.

## Types of Attention Mechanisms

There are several types of attention mechanisms that have been proposed in the literature, with the main variants being global attention and local attention.

### Global Attention

Global attention, also known as soft attention, considers all the hidden states of the input sequence when computing the context vector for each output time step. The model learns to assign a weight to each input state, which determines the degree of "attention" the output should pay to each input element.

### Local Attention

Local attention, or hard attention, is a more computationally efficient variant where the model only considers a subset of the input states within a window around a certain position. This approach reduces the computational complexity since it does not require attending to the entire input sequence at each decoding step.

![global-vs-local-attention](https://www.kdnuggets.com/wp-content/uploads/Fig5-Chauhan-attention-mechanism-deep-learning-explained.jpg)

Attention mechanisms have become a fundamental component in modern NLP architectures, leading to the development of highly successful models like the Transformer, which relies entirely on self-attention to process sequences. The flexibility and effectiveness of attention have made it a standard tool in the NLP practitioner's toolkit.

# Transformer Architecture

## Breakdown of the Transformer Model

The Transformer model, proposed in the paper "Attention is All You Need" by Vaswani et al., represents a significant departure from previous sequence-to-sequence models that relied on RNNs or LSTMs. The Transformer model is based entirely on attention mechanisms, specifically the self-attention mechanism, without any recurrent architecture.

### Model Architecture

![transformer](https://deeprevision.github.io/posts/001-transformer/transformer.png)

The Transformer follows an encoder-decoder structure. Here's a high-level breakdown of its components:

- **Encoder**: The encoder maps an input sequence of symbol representations to a sequence of continuous representations. The encoder is composed of a stack of identical layers, each containing two sub-layers: a multi-head self-attention mechanism and a position-wise fully connected feed-forward network.

- **Decoder**: The decoder is also composed of a stack of identical layers. In addition to the two sub-layers in each encoder layer, the decoder inserts a third sub-layer, which performs multi-head attention over the output of the encoder stack.

### Key Components

- **Self-Attention**: This mechanism allows the model to weigh the influence of different parts of the input sequence when encoding a particular element. Self-attention is computed using the scaled dot-product attention formula:

$$ \text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V $$

where $ Q $, $ K $, and $ V $ are the queries, keys, and values, respectively, and $ d_k $ is the dimensionality of the keys.

- **Positional Encoding**: Since the model does not contain recurrence or convolution, positional encodings are added to the input embeddings to provide information about the position of the tokens in the sequence. The positional encodings have the same dimension as the embeddings so that the two can be summed up. They are defined as:

$$ PE_{(pos, 2i)} = \sin\left(\frac{pos}{10000^{2i/d_{\text{model}}}}\right) $$
$$ PE_{(pos, 2i+1)} = \cos\left(\frac{pos}{10000^{2i/d_{\text{model}}}}\right) $$

where $ pos $ is the position, $ i $ is the dimension, and $ d_{\text{model}} $ is the dimensionality of the model's embeddings.

![positional-encoding](https://machinelearningmastery.com/wp-content/uploads/2022/01/PE3.png)

- **Multi-Head Attention**: The model uses multiple heads for the attention mechanism to allow the model to focus on different positions. This is achieved by projecting the queries, keys, and values multiple times with different, learned linear projections.

![multi-head](https://production-media.paperswithcode.com/methods/multi-head-attention_l1A3G7a.png)

# Understanding BERT and GPT

Two Transformer-based models have stood out for their effectiveness and versatility: BERT (Bidirectional Encoder Representations from Transformers) and GPT (Generative Pretrained Transformer). Both models have pushed the boundaries of what's possible in NLP, achieving state-of-the-art results on a range of tasks.

## BERT (Bidirectional Encoder Representations from Transformers)

### Introduction to BERT

BERT is a transformer-based model designed to understand the context of a word in search queries and other text by looking at the words that come before and after it. This is known as bidirectional context, which is a key differentiator between BERT and GPT.

### Architecture of BERT

BERT is based on the Transformer architecture, which uses self-attention mechanisms. It consists of multiple layers of bidirectional transformers. BERT comes in two sizes, BERT Base and BERT Large, with the latter having more transformer blocks (layers), more attention heads, and a larger hidden layer size.

![bert](https://www.aionlinecourse.com/uploads/blog/image/BERT1.png)

### Pre-training and Fine-tuning

BERT is pre-trained on a large corpus of text in two unsupervised tasks: Masked Language Modeling (MLM) and Next Sentence Prediction (NSP). In MLM, some percentage of the input tokens are masked at random, and then the model attempts to predict the masked words, based on their context. In NSP, the model learns to predict whether two sentences are likely to be consecutive.

![mlm](https://raw.githubusercontent.com/UKPLab/sentence-transformers/master/docs/img/MLM.png)

![bert](https://production-media.paperswithcode.com/methods/new_BERT_Overall.jpg)

After pre-training, BERT can be fine-tuned with additional output layers for a wide range of tasks without substantial task-specific architecture modifications.

### Applications of BERT

BERT has been successfully applied to many NLP tasks, including:

- Text classification
- Question answering
- Named entity recognition
- Sentiment analysis

## GPT (Generative Pretrained Transformer)

### Introduction to GPT

GPT, like BERT, is based on the Transformer model, but unlike BERT, GPT is unidirectional and is designed to generate text. GPT takes an input of text and continues to generate text based on this input.

![bert-vs-gpt](https://www.researchgate.net/publication/354908597/figure/fig1/AS:1073191841722368@1632880282219/Model-structure-of-BERT-and-GPT.png)

### Architecture of GPT

The architecture of GPT is similar to the decoder part of the Transformer model. It uses masked self-attention, where each token can only attend to previous tokens in the self-attention layers of the transformer.

![gpt](https://production-media.paperswithcode.com/methods/Screen_Shot_2020-05-27_at_12.41.44_PM.png)

### Pre-training and Fine-tuning

GPT is also pre-trained on a large corpus of text in an unsupervised manner. It uses a task called Language Modeling, where the model predicts the next word in a sentence given the previous words. After pre-training, GPT can be fine-tuned on a specific task by adding an additional output layer and continuing the training process on a labeled dataset.

### Applications of GPT

GPT has been used for applications such as:

- Text generation
- Conversational agents
- Language translation
- Text summarization

Both BERT and GPT have their strengths and are chosen based on the requirements of the task. BERT's bidirectional context is powerful for understanding the meaning of words within their context, making it ideal for tasks that require understanding the relationship between words. GPT's strength lies in generating coherent and contextually relevant text, making it suitable for tasks that involve text continuation.

These models have paved the way for more advanced language models like GPT-2, GPT-3, and BERT-based models like RoBERTa and DistilBERT, each offering improvements and variations that make them suited for different NLP tasks.

```python
from datasets import load_dataset

# Load the IMDB dataset
dataset = load_dataset("imdb")

# Split the dataset into training and testing sets
train_dataset = dataset["train"]
test_dataset = dataset["test"]
```

```python
train_dataset[0]
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the dataset
def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True, padding=True, max_length=128)

tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True)
tokenized_test_dataset = test_dataset.map(preprocess_function, batched=True)
```

```python
# Convert the labels into binary format
def convert_labels(example):
    if example['label'] == 1:
        example['label'] = 1  # Positive sentiment
    else:
        example['label'] = 0  # Negative sentiment
    return example

binary_train_dataset = tokenized_train_dataset.map(convert_labels)
binary_test_dataset = tokenized_test_dataset.map(convert_labels)
```

```python
# Load the pre-trained BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
```

```python
# Define the training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir="./logs",
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=binary_train_dataset,
    eval_dataset=binary_test_dataset,
)

# Train the model
trainer.train()
```

```python
# Evaluate the model
results = trainer.evaluate()
print(results)
```

```python
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128).to(device)
    outputs = model(**inputs)
    prediction = torch.argmax(outputs.logits, dim=-1)
    return "positive" if prediction.item() == 1 else "negative"
```

```python
# Test on a new review
new_review = "The movie was fantastic! I loved it."
print(predict_sentiment(new_review))
```

```python
# Test on another review
new_review = "The movie was not very great."
print(predict_sentiment(new_review))
```


## Named Entity Recognition (NER) with BERT

Named Entity Recognition (NER) is a subtask of information extraction that seeks to locate and classify named entities mentioned in unstructured text into predefined categories such as the names of persons, organizations, locations, expressions of times, quantities, monetary values, percentages, etc.

For example, in the sentence "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in Cupertino, California, on April 1, 1976," a NER system would identify "Apple Inc." as an organization, "Steve Jobs," "Steve Wozniak," and "Ronald Wayne" as persons, "Cupertino, California" as a location, and "April 1, 1976" as a date.

We'll use pre-trained BERT model for Named Entity Recognition (NER).

```python
from transformers import AutoModelForTokenClassification, AutoTokenizer
```

```python
# Load a pre-trained model and tokenizer
model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
model = AutoModelForTokenClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Encode text
text = "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very close to the Manhattan Bridge."
inputs = tokenizer.encode(text, return_tensors="pt")
```

```python
# Predict NER tags
with torch.no_grad():
    outputs = model(inputs).logits
predictions = torch.argmax(outputs, dim=2)
```

```python
# Convert predictions to labels
labels = [model.config.id2label[prediction] for prediction in predictions[0].tolist()]
```

```python
# Print the tokens with their predicted NER tags
for token, label in zip(tokenizer.tokenize(text), labels):
    print(f"{token}: {label}")
```

## Text Generation with GPT-2

Let's use a pre-trained GPT-2 model to generate text. Hugging Face provides several versions of GPT-2 that are already trained on a diverse corpus of text data.

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
# Load pre-trained GPT-2 model and tokenizer
model_name = 'gpt2'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
def preprocess(text):
    # Encode the input text
    return tokenizer.encode(text, return_tensors='pt', add_special_tokens=True)

def generate(text, max_length=150):
    # Preprocess the text
    input_ids = preprocess(text)

    # Generate
    summary_ids = model.generate(input_ids,
                                 max_length=max_length,
                                 num_beams=5,
                                 no_repeat_ngram_size=2,
                                 early_stopping=True)

    # Decode the text
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
text_to_generate = """
The history of natural language processing (NLP) generally started in the 1950s, although work can be found from earlier periods.
In 1950, Alan Turing published an article titled "Computing Machinery and Intelligence" which proposed what is now called the Turing test as a criterion of intelligence.
"""

generated = generate(text_to_generate)
print(generated)
```
